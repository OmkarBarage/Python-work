```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np
```

```python
df["Year"] = df["Date"].dt.year
df = df[df["Year"] == 2019]
```

**To find total sale we need Total price , i.e  by monthly
 for Getting Total Price  = Price * Quantity
 and combine data with month**
 

```python
df['Total_Price'] = df['Price']*df['Quantity']

df['Month'] = df['Date'].dt.month
```

```python
df
```

```python
#df.set_index('Date', inplace=True)
#monthly_sales_s = df.resample('M').sum()
```

```python
# group the data by month and calculate the total sales for each month
monthly_sales = df.groupby('Month')['Total_Price'].sum()
```

```python
monthly_sales
```

```python
# create line plot using seaborn
sns.set(style="darkgrid")
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, palette="hue1")

# add labels and title
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.title("Monthly Sales Trend")

plt.yticks(monthly_sales.values, [f"{x:,}" for x in monthly_sales.values])

plt.show()

```

```python
sns.set(style="darkgrid")
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values)

# Add data labels to the plot
for x,y in zip(monthly_sales.index, monthly_sales.values):
    label = "{:.2f}".format(y)
    plt.annotate(label, (x,y), textcoords="offset points", xytext=(0,10), ha='center')

# Add labels and title
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.title("Monthly Sales Trend")

# Display the plot
plt.show()
```

**2.What are the most frequently purchased products (Top 10)?
**

```python
top_products = df.groupby('ProductName')['Quantity'].sum().sort_values(ascending=False)[:10]
print(top_products)
```

```python
sns.set(style="darkgrid")

# group data by product name and sum the quantity
top_products = df.groupby('ProductName')['Quantity'].sum().sort_values(ascending=False)[:10]

# create bar plot
plt.figure(figsize=(10,6))
sns.barplot(x=top_products.values, y=top_products.index, palette="rocket")

# add labels and title
plt.xlabel("Total Quantity")
plt.ylabel("Product Name")
plt.title("Top 10 Most Frequently Purchased Products")

plt.show()
```

**3.How many products does the customer purchase in each transaction (Avg.)?**

```python
avg_products_per_transaction = df.groupby('TransactionNo')['ProductName'].nunique()
print(avg_products_per_transaction)
```

**4.Top ten countries by there sales and quantity of products.**

```python
 top_countries = df.groupby('Country')[['Total_Price','Quantity']].sum().sort_values(['Total_Price','Quantity'], ascending=False)[:10]
print(top_countries)
```

```python
# create bar plot using seaborn
sns.set(style="darkgrid")
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.index, y="Total_Price", data=top_countries)

# add labels and title
plt.xlabel("Country")
plt.ylabel("Total Sales")
plt.title("Top 10 Countries by Total Sales")

 

# add value labels on top of each bar
for i, v in enumerate(top_countries['Total_Price']):
    plt.text(i, v + 10000, f"{v:,.0f}", ha='center')

plt.show()
```

**5.Countries with their customer base (Top 10)
**

```python
top_countries_customers = df.groupby('Country')['TransactionNo'].nunique().sort_values(ascending=False)[:10]
print(top_countries_customers)
```

```python
# create bar plot using seaborn
sns.set_style("whitegrid")
plt.figure(figsize=(10,6))
sns.barplot(x=top_countries_customers.index, y=top_countries_customers.values)
plt.xlabel("Country")
plt.ylabel("Number of Unique Transactions")
plt.title("Top 10 Countries by Customer Base")
plt.show()
```
